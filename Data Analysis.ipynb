```python
import pandas as pd
import matplotlib as plt

df = pd.read_excel(r'DataModel-PizzaSales.xlsx')


#This will delete the FIRST and LAST letters in 'pizza_name', (these are simply the words 'The' and 'Pizza')
df['pizza_name'] = df['pizza_name'].str.split().apply(lambda x: ' '.join(x[1:-1]))

#This will delete the YEAR portion in order_date. (From YYYY-MM-DD --> MM-DD). All data is from 2015
#df['order_date'] = df['order_date'].dt.strftime('%m-%d')
```

```python
### HISTOGRAM + Scatter Plot ### 
'''
Histogram: df.plot.hist()
Scatter Plot:df.plot.scatter(x=’column1′,y=’column2′)
'''

#Counts the pizza_ids
pizzaIDCounts = df['pizza_id'].value_counts()
#Counts the pizza names
pizzaNameCounts = df ['pizza_name'].value_counts()
```

```python
pizzaIDCounts, pizzaNameCounts
```

Already, there is a very obvious difference in the most popular order (big_meat_s, 1811) and least popular order(the_greek_xxl, 28) of pizza. The specific type of pizza + size combo is nice to know, and we'll dig a little deeper later.


Digging a little deeper, does the size of the pizza matter?: There is a significant (couple thousand) difference in size L pizzas and M/S/XL/XXL

```python
sizeCounts = df['pizza_size'].value_counts()
print(sizeCounts.head())
sizeCounts.plot.bar(xlabel = "pizza_size")
```

There is a clear difference between what size pizza is the most popular - Large pizzas are super popular! But XL and XXL are rarely ever ordered (544, 28).


If Large is by far the most popular size of pizza to order, it might also be good to know what type of pizza is also the most popular. Looking at the below graph, the Classic Deluxe, is the most popular with a close 6 right behind it. However, there Brie Carre pizza is by far the least popular with about only 500 orders throughout the year.

```python
groupedByDate = df.groupby('order_date')
dailyProfits = groupedByDate.agg({'total_price':'sum'})

dailyProfits
```

```python
dailyProfits.plot.bar()
```

The above shows all profits in 2015. It seems like we are missing about a week's worth of data as we only have 358 rows and there are 365 days in a year. This is most likely because the pizza place could be closed on major holidays.

We can tell that there are some days that are significantly more profitable than others, but it's really hard to tell what exact days those are on. We could probably look them up, but it might also be nice to know what months are more profitable than others, and we can go from there.

```python
groupedByMonth = df.groupby(df['order_date'].dt.month)
monthlySales = groupedByMonth['total_price'].sum()

monthlySalesPD = pd.DataFrame(monthlySales)
monthlySalesPD['Months'] = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

monthlySalesPD
```

```python
monthlySalesPD.plot.bar(x = 'Months', y = 'total_price', color = 'pink')
```

This is a simple bar graph that displays the profits for the month. Depending on who you ask, (Whether ~$5000 is "a lot" or not) one might be able to say that the profits from month to month are fairly consistent/fluctuate. Personally, I'd say that these are rather consistent and is a good sign of a good business!
